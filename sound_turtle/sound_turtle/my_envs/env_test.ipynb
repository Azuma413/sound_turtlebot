```python
import numpy as np
import matplotlib.pyplot as plt
from my_env import MyEnv, WrapDrQ
import cv2

room_num = 0
env1 = MyEnv(room_num)
```

```python
_ = env1.reset()
```

```python
def get_image(room_num):
    env1 = MyEnv(room_num)
    obs1 = env1.reset() # 環境のリセット
    obs = np.zeros_like(obs1)
    obs[:,:,0] = obs1[:,:,1]
    obs[:,:,1] = obs1[:,:,1]
    obs[:,:,2] = obs1[:,:,1]
    obs2 = env1.image_list[0]
    # 480*480*3を128*128*3にリサイズ
    obs2 = cv2.resize(obs2, (128, 128))
    obs[:,:,0] = obs2[:,:,2]
    obs[:,:,1] = obs2[:,:,2]
    return obs

image_list = []
for i in range(4):
    obs = get_image(i)
    image_list.append(obs)
fig, ax = plt.subplots(1, 4, figsize=(12, 3))
for i in range(4):
    ax[i].imshow(image_list[i])
    ax[i].axis('off')
plt.show()
```
